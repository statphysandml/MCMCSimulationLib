# Error comparison

```python
from pystatplottools.pdf_env.loading_figure_mode import loading_figure_mode
fma, plt = loading_figure_mode(develop=True)

import numpy as np
import pandas as pd

from mcmctools.utils.utils import linspace_rp_intervals
from mcmctools.mcmc.evaluation_module import EvaluationModule
```

```python
evaluation_module = EvaluationModule(sim_base_dir="../", rel_data_dir="data/ScalarTheorySimulation/", rel_results_dir="results/ScalarTheorySimulation/",
                                     running_parameter_kind=None,
                                     running_parameter="kappa",
                                     rp_values=linspace_rp_intervals(0.22, 0.3, 9))
```

```python
evaluation_module.compute_expectation_values(measures=["Mean", "AbsMean", "SecondMoment", "Action"], error_type="statistical")
statistical_error = evaluation_module.computed_expectation_values
evaluation_module.compute_expectation_values(measures=["Mean", "AbsMean", "SecondMoment", "Action"], error_type="bootstrap", n_means_bootstrap=200)
bootstrap_error = evaluation_module.computed_expectation_values
evaluation_module.compute_expectation_values(measures=["Mean", "AbsMean", "SecondMoment", "Action"], error_type="jackknife")
jackknife_error = evaluation_module.computed_expectation_values
```

```python
statistical_error
```

```python
fig, ax =fma.newfig(1, figsize=(10, 5))
ax.plot(statistical_error["Std"]["Mean"], "-", label="Statistical", lw=3)
ax.plot(bootstrap_error["Std"]["Mean"], "--", label="Bootstrap", lw=3)
ax.plot(jackknife_error["Std"]["Mean"], "-.", label="Jackknife", lw=3)
ax.legend()
plt.show()
```

```python
fig, ax =fma.newfig(1, figsize=(10, 5))
ax.plot(statistical_error["Std"]["AbsMean"], "-", label="Statistical", lw=3)
ax.plot(bootstrap_error["Std"]["AbsMean"], "--", label="Bootstrap", lw=3)
ax.plot(jackknife_error["Std"]["AbsMean"], "-.", label="Jackknife", lw=3)
ax.legend()
plt.show()
```

```python
fig, ax =fma.newfig(1, figsize=(10, 5))
ax.plot(statistical_error["Std"]["SecondMoment"], "-", label="Statistical", lw=3)
ax.plot(bootstrap_error["Std"]["SecondMoment"], "--", label="Bootstrap", lw=3)
ax.plot(jackknife_error["Std"]["SecondMoment"], "-.", label="Jackknife", lw=3)
ax.legend()
plt.show()
```

```python
fig, ax =fma.newfig(1, figsize=(10, 5))
ax.plot(statistical_error["Std"]["Action"], "-", label="Statistical", lw=3)
ax.plot(bootstrap_error["Std"]["Action"], "--", label="Bootstrap", lw=3)
ax.plot(jackknife_error["Std"]["Action"], "-.", label="Jackknife", lw=3)
ax.legend()
plt.show()
```
